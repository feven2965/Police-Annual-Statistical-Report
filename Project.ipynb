```python
import pandas as pd
import numpy as np

# Read the CSV file into a DataFrame
df = pd.read_csv('/content/Calls for Service Attended.csv')

# Calculate Z-score for 'EVENT_COUNT' column
df['EVENT_COUNT_ZSCORE'] = np.abs((df['EVENT_COUNT'] - df['EVENT_COUNT'].mean()) / df['EVENT_COUNT'].std())

# Define a threshold for Z-score (e.g., 3)
zscore_threshold = 3

# Identify outliers based on the threshold
outliers_df = df[df['EVENT_COUNT_ZSCORE'] > zscore_threshold]

# Print or further analyze the outliers
print(outliers_df)
```

```python
# Count occurrences of each neighborhood
neighborhood_counts = df['NEIGHBOURHOOD_158'].value_counts()

# Print the counts
print(neighborhood_counts)
```

```python
# Group by neighborhood and calculate mean, median, etc. for EVENT_COUNT
neighborhood_summary = df.groupby('NEIGHBOURHOOD_158')['EVENT_COUNT'].agg(['mean', 'median', 'std'])

# Print the summary statistics
print(neighborhood_summary)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.read_csv('Calls for Service Attended.csv')

# Select numerical columns for correlation analysis
numerical_columns = ['_id', 'EVENT_YEAR', 'EVENT_COUNT']

# Compute correlation matrix
correlation_matrix = data[numerical_columns].corr()

# Plot heatmap of correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose


data = pd.read_csv('Calls for Service Attended.csv')

# Set a numerical index for time series analysis
data.index = pd.RangeIndex(start=0, stop=len(data))

# Univariate Time Series Analysis

# Plot time series of 'EVENT_COUNT' with numerical index as time
plt.figure(figsize=(10, 6))
plt.plot(data['EVENT_COUNT'])
plt.xlabel('Time')
plt.ylabel('Event Count')
plt.title('Time Series of Event Count')
plt.grid(True)
plt.show()

# Seasonal Decomposition

# Decompose time series into trend, seasonal, and residual components
decomposition = seasonal_decompose(data['EVENT_COUNT'], period=12)

# Plot decomposition components
plt.figure(figsize=(10, 8))
plt.subplot(411)
plt.plot(data['EVENT_COUNT'], label='Original')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(decomposition.trend, label='Trend')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonal')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(decomposition.resid, label='Residual')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()

# Forecasting

from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA model to the data
model = ARIMA(data['EVENT_COUNT'], order=(1, 1, 1))  # Example order for ARIMA model
model_fit = model.fit()

# Forecast future values
forecast_values = model_fit.forecast(steps=30)  # Forecasting 30 steps ahead

# Plot original data and forecasted values
plt.figure(figsize=(10, 6))
plt.plot(data['EVENT_COUNT'], label='Original')
plt.plot(range(len(data), len(data) + 30), forecast_values, label='Forecast', color='red')
plt.xlabel('Time')
plt.ylabel('Event Count')
plt.title('Time Series Forecasting')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load your police data into a pandas DataFrame
data = pd.read_csv('Calls for Service Attended.csv')

# Set a numerical index for time series analysis
data.index = pd.RangeIndex(start=2015, stop=2015+len(data), step=1)

# Univariate Time Series Analysis

# Plot time series of 'EVENT_COUNT' with numerical index as time and set x-axis ticks to match years
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['EVENT_COUNT'])
plt.xlabel('Year')
plt.ylabel('Event Count')
plt.title('Time Series of Event Count')
plt.xticks(data.index, rotation=45)
plt.grid(True)
plt.show()

# Seasonal Decomposition

# Decompose time series into trend, seasonal, and residual components
decomposition = seasonal_decompose(data['EVENT_COUNT'], period=12)

# Plot decomposition components
plt.figure(figsize=(10, 8))
plt.subplot(411)
plt.plot(data.index, data['EVENT_COUNT'], label='Original')
plt.legend(loc='upper left')
plt.xticks(data.index, rotation=45)
plt.subplot(412)
plt.plot(data.index, decomposition.trend, label='Trend')
plt.legend(loc='upper left')
plt.xticks(data.index, rotation=45)
plt.subplot(413)
plt.plot(data.index, decomposition.seasonal, label='Seasonal')
plt.legend(loc='upper left')
plt.xticks(data.index, rotation=45)
plt.subplot(414)
plt.plot(data.index, decomposition.resid, label='Residual')
plt.legend(loc='upper left')
plt.xticks(data.index, rotation=45)
plt.tight_layout()
plt.show()
```
